```python
class Property:
    def __init__(self, price):
        self.price = price

    def calculate_stamp_duty(self):
        """Calculate stamp duty based on the property price."""
        if self.price <= 300000:
            return self.price * 0.01
        elif self.price <= 2000000:
            return self.price * 0.03
        elif self.price <= 3000000:
            return self.price * 0.04
        else:
            return self.price * 0.05  # Simplified rates for demonstration

    def calculate_second_mortgage(self, mortgage_percentage):
        """Calculate the amount required for the second mortgage if necessary."""
        if mortgage_percentage > 70:
            first_mortgage = self.price * (mortgage_percentage / 100)
            second_mortgage_needed = self.price - first_mortgage
            return second_mortgage_needed
        return 0

class Mortgage:
    def __init__(self, income, repayment_period, interest_rate):
        self.income = income
        self.repayment_period = repayment_period  # in years
        self.interest_rate = interest_rate / 100  # Convert to percentage

    def calculate_monthly_repayment(self, loan_amount):
        """Calculate the monthly repayment for a given loan amount."""
        monthly_interest_rate = self.interest_rate / 12
        number_of_payments = self.repayment_period * 12
        if monthly_interest_rate == 0:
            return loan_amount / number_of_payments
        return (loan_amount * monthly_interest_rate) / (1 - (1 + monthly_interest_rate) ** -number_of_payments)

def main():
    # Input property details
    property_price = float(input("Enter the property price (in HKD): "))
    mortgage_percentage = float(input("Enter the mortgage percentage (e.g. 70 for 70%): "))

    property = Property(property_price)

    # Calculate stamp duty
    stamp_duty = property.calculate_stamp_duty()
    print(f"Stamp Duty to be paid: HKD {stamp_duty:.2f}")

    # Calculate first mortgage amount
    first_mortgage = property_price * (mortgage_percentage / 100)
    print(f"First Mortgage Amount: HKD {first_mortgage:.2f}")

    # Calculate second mortgage if necessary
    second_mortgage = property.calculate_second_mortgage(mortgage_percentage)
    if second_mortgage > 0:
        print(f"Second Mortgage Amount Needed: HKD {second_mortgage:.2f}")

        # Input additional details for second mortgage
        income = float(input("Enter your monthly income (in HKD): "))
        repayment_period = int(input("Enter the repayment period (in years): "))
        interest_rate = float(input("Enter the interest rate (as a percentage): "))

        mortgage = Mortgage(income, repayment_period, interest_rate)
        monthly_expense = mortgage.calculate_monthly_repayment(second_mortgage)
        print(f"Monthly Expense for Second Mortgage: HKD {monthly_expense:.2f}")

    # Calculate total monthly housing expense
    total_monthly_repayment = (first_mortgage / (repayment_period * 12)) + \
                               (monthly_expense if second_mortgage > 0 else 0)

    print(f"Total Monthly Housing Expense: HKD {total_monthly_repayment:.2f}")

if __name__ == "__main__":
    main()
```
